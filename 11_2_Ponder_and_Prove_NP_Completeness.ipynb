### 16.7.2.2

## Question reworded

If we have a set of undirected graphs, with a set of nodes $N$ and a set of edges $E \subseteq N * N$, and we try to bi-color the graphs, is it possible to do a breadth-first search?

## Answer

We would argue yes. We start with the first node, and we assign all the connecting nodes to the opposite color of node 1. From there we move to node 2, and check if its connecting nodes are appropiately colored. If they are not, then the process must restart with a different color scheme.

Since we are only using two colors, this will work with such a method. This would result in a performance of $O(2^n)$. The key to this method working is having the number of colors be small. If we bumped up our colors to let's say 50, this would not be a practically computable concept.

### 16.7.2.8

## Reword Question

In a 5-clique graph (also known as a K5 graph), how many Hamiltonian cycles are there? How many are there for a Kn graph?

## Part 1 Answer

Our answer is 12

Originally we thought the answer was 4! * 5 which is 120. We thought this because if we start with a node, there are 4 choices. Once a choice is made, we are left with 3 choices and so on. Then we figured we would times our result by 5 because there are five nodes.

However, after some research, we learned that to discover unique Hamiltonian cycles, we need to find uniquely shaped paths. This would reduce our answer down significantly.  We would need to divide our answer by two, since every path has a reverse (which would amount to the same shape). Next, we would have to divide our answer by 5 since it does not matter which node we start one (they will produce the same shape).

This results in our answer of 12

4! * 5 = 120

120 / 2 = 60

120 / 5 = 12

12

## Part 2 Answer

We can reduce our answer from part 1 to a formula:

$5!/(5*2)$

If we were to determine how many distinct Hamiltonian cycles were in a Kn graph, we would apply this formula with n taking the place of 5.


Since originally our formula looked like $n!/n*2$, we can reduce this down to $(n-1)!/2$

We divide by 2, since every Hamiltonian cycle, has a reverse (making it not distinct).

$(n-1)!/2$

$$Decider_{HalfClique_{TM}}(Map(<G, k>))= \left\{
\begin{array}{il}
\text{accepts} \rightarrow Map(<G, k>) \text{ returns G} \rightarrow\\
 \text{G is a clique }\\
 \rightarrow Clique \text{ contains } <G, k> w\\
\\
\text{rejects} \rightarrow Map(<G, k>) \text{ returns G'} \rightarrow\\
 \text{G' is not a clique}\\
 \rightarrow Clique \text{ does NOT contain } <G, k>\\
\end{array}
\right.$$

## Explanation

This is a mapping reduction from *Clique* to *Half-Clique*.
We show that to find a *Half-Clique*, we are dependant on the isClique function.

Therefore, if *Clique* is NPC, then *Half-Clique* is also NPC.

## Note

We used 4 in

```G' = a set of 4 unconnected nodes```

Because of the following:

* The number should be even to be divisible cleanly by 2

* If we used 2, that would result in a graph with 1 node, which could be argued as a clique

* By default, 4 is the smallest number to achieve the results we aimed for.

* With 4, it is cleanly divisble by 2, and a graph with 2 unconnected nodes is most definitely not a clique.

````
import pycosat

N = 9
M = 3

def exactly_one(variables):
    cnf = [ variables ]
    n = len(variables)

    for i in range(n):
        for j in range(i+1, n):
            v1 = variables[i]
            v2 = variables[j]
            cnf.append([-v1, -v2])

    return cnf

def transform(i, j, k):
    return i*N*N + j*N + k + 1

def inverse_transform(v):
    v, k = divmod(v-1, N)
    v, j = divmod(v, N)
    v, i = divmod(v, N)
    return i, j, k

if __name__ == '__main__':
    cnf = []

    # Cell, row and column constraints
    for i in range(N):
        for s in range(N):
            cnf = cnf + exactly_one([ transform(i, j, s) for j in range(N) ])
            cnf = cnf + exactly_one([ transform(j, i, s) for j in range(N) ])

        for j in range(N):
            cnf = cnf + exactly_one([ transform(i, j, k) for k in range(N) ])

    # Sub-matrix constraints
    for k in range(N):
        for x in range(M):
            for y in range(M):
                v = [ transform(y*M + i, x*M + j, k) for i in range(M) for j in range(M)]
                cnf = cnf + exactly_one(v)

    # See contribution from @GregoryMorse below
    cnf = { frozenset(x) for x in cnf }
    cnf = list(cnf)

    # A 16-constraint Sudoku
    constraints = [
        (0, 3, 7),
        (1, 0, 1),
        (2, 3, 4),
        (2, 4, 3),
        (2, 6, 2),
        (3, 8, 6), 
        (4, 3, 5),
        (4, 5, 9),
        (5, 6, 4),
        (5, 7, 1),
        (5, 8, 8),
        (6, 4, 8),
        (6, 5, 1),
        (7, 2, 8),#change 2 to 8
        (7, 7, 5), 
        (8, 1, 4),
        (8, 6, 3),
    ]

    cnf = cnf + [[transform(z[0], z[1], z[2])-1] for z in constraints]

    for solution in pycosat.itersolve(cnf):
        X = [ inverse_transform(v) for v in solution if v > 0]
        for i, cell in enumerate(sorted(X, key=lambda h: h[0] * N*N + h[1] * N)):
            print(cell[2]+1, end=" ")
            if (i+1) % N == 0: print("")

        print('')
````
